```python
import cv2 as cv
import numpy as np
```

```python
def empty(a):
    pass

cv.namedWindow('HSV')#create a slider window
cv.resizeWindow('HSV', 500, 200)#set the size of the window
cv.createTrackbar('HUE min','HSV',0,179,empty)#slider name, window in which you want it, start position, max length, function that needs to be called everytime the trackbar event is called
cv.createTrackbar('HUE max','HSV',179,179,empty)
cv.createTrackbar('Saturation min','HSV',0,255,empty)
cv.createTrackbar('Saturation max','HSV',255,255,empty)
cv.createTrackbar('Value min','HSV',0,255,empty)
cv.createTrackbar('Value max','HSV',255,255,empty)

# frameWidth = 640
# frameHeight = 480
cap = cv.VideoCapture(1)#initialise the external webcam
# cap.set(3, frameWidth)
# cap.set(4, frameHeight)

while True:
    _,img = cap.read()
    hsvimg = cv.cvtColor(img, cv.COLOR_BGR2HSV) #convert bgr colour space to hsv colour space
#     img = cv.resize(img, (400,400))
    h_min = cv.getTrackbarPos('HUE min', 'HSV')#Get the position of the hsv values
    h_max = cv.getTrackbarPos('HUE max', 'HSV')
    s_min = cv.getTrackbarPos('Saturation min', 'HSV')
    s_max = cv.getTrackbarPos('Saturation max', 'HSV')
    v_min = cv.getTrackbarPos('Value min', 'HSV')
    v_max = cv.getTrackbarPos('Value max', 'HSV')
    print(h_min,s_min, v_min)
    print(h_max, s_max, v_max)
    
    lower = np.array([h_min,s_min,v_min])
    upper = np.array([h_max,s_max,v_max])
    mask = cv.inRange(hsvimg, lower,upper) #return only what lies in this range
    result = cv.bitwise_and(img,img,mask=mask)
    
    mask = cv.cvtColor(mask, cv.COLOR_GRAY2BGR)#mask image will be black and white. hence, just one channel.Cant merge an image with one channel to an image with 3 channels.
    hstack = np.hstack([img,mask,result])#horizontally stack the images such that they appear side by side
    hstack = cv.resize(hstack, (800,300))#resizing the output window
    
#     cv.imshow("frame", frame)
#     cv.imshow('HSVIMG', hsvimg)
    cv.imshow('Hstack', hstack)
    if cv.waitKey(1) == 27:
        break
        
cap.release()
cv.destroyAllWindows()
```
